```python
G = nx.Graph()

G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)

G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(1, 4)
```

```python
# random graph
G1 = nx.erdos_renyi_graph(25, 0.15)

# visualize graph
fig, ax = plt.subplots(figsize=(3, 2),dpi=300)
fig.patch.set_linewidth(1)
fig.patch.set_edgecolor('black')
nx.draw(G1, with_labels=True, node_size=25,width=0.2, font_size=3)
```

```python
number_of_nodes = nx.number_of_nodes(G1)
print("Total number of nodes: ",number_of_nodes)
```

```python
number_of_edges = nx.number_of_edges(G1)
print("Total number of edges: ",number_of_edges)
```

```python
degree = dict(nx.degree(G1))
print("Type of variable: ",type(degree))
print("Degree of node 23: ",degree[23])
```

```python
# create an instance of graph
directedG = nx.DiGraph()

# add nodes from a list
directedG.add_nodes_from([1, 2, 3, 4])

# add edges from a list
directedG.add_edges_from([(1, 2), (1, 3), (1, 4)])

# visualize graph
plt.figure()
nx.draw(directedG, with_labels=True, node_size=1000, font_size=30, arrowsize=30)
```

```python
print("The graph is directed: ",nx.is_directed(directedG))
print("The graph is weighted: ",nx.is_weighted(G1))
```

```python
# create edgelist
edgelist = np.array([(0, 1), (0, 2), (0, 3)])

# graph using edgelist
G = nx.from_edgelist(edgelist)

# visualize graph
plt.figure(figsize=(10, 6))
nx.draw(G, with_labels=True, node_size=3000, font_size=40)
```

```python
from itertools import combinations

# adjacency matrix
adj = [[0,1,1,1],[1,0,1,1],[1,1,0,1],[1,1,1,0]]

# create graph instance
G = nx.Graph(np.array(adj))

# check if G is fully connected
def is_fully_connected(G):
    for (i,j) in combinations(G.nodes(),2): 
        if not G.has_edge(i,j):
            print("Graph is NOT fully connected!")
    print("Graph is fully connected!")

is_fully_connected(G)
```
